```python
import pandas as pd
```

```python
sales=pd.read_csv('C:/Users/hp/Downloads/Sales.csv',encoding='ISO-8859-1')
sales
```

```python
sales.dtypes
```

```python
date_formats=['%d/%m/%Y','%m/%d/%Y','%d-%m-%Y','%m-%d-%Y']

def parse_date(date_str):
    for fmt in date_formats:
        try:
            return pd.to_datetime(date_str,format=fmt)
        except ValueError:
            continue
    return pd.NaT
   
# Apply parsing function to the date column
sales['Order Date']=sales['Order Date'].apply(parse_date)
sales['Delivery Date']=sales['Delivery Date'].apply(parse_date)


# Convert to the desired format
#sales['Order Date']=sales['Order Date'].astype(str)
#sales['Delivery Date']=sales['Delivery Date'].astype(str)
```

```python
# Convert to the desired format
sales['Order Date']=sales['Order Date'].dt.strftime('%Y-%m-%d')
sales['Delivery Date']=sales['Delivery Date'].dt.strftime('%Y-%m-%d')
```

```python
sales.isnull().sum()
```

```python
sales.fillna('NaT',inplace=True)
```

```python
sales
```
